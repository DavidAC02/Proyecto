```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH_TRAIN = "Sellos/" 
train = []
classes = []
for folder in os.listdir(PATH_TRAIN):
    for filename in os.listdir(PATH_TRAIN + folder):
        filename = PATH_TRAIN+folder+"/" +filename
        img = cv2.imread(filename)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        img = cv2.resize(img,(80,80))
        train.append([img])
        classes.append(folder)
#train = np.array(train, dtype='object')
#classes = np.array(classes, dtype='<U9')
```

```python
np.save(os.path.join('','features'),np.array(train))
```
